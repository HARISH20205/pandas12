```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.linspace(0,12,2000)
y=np.sin(x)
```

```python
plt.figure(figsize=(4,3),dpi=200)
plt.plot(x,y,'--g',linewidth=3)

#plt.xlim(4,5)
#or plt.axis([4,5,0.1,0.9])

#if everything is to fit in axis then 
plt.axis('tight')
```

```python
x=np.linspace(0,13,100)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,'-.r',label='Sin(x)')
plt.plot(x,y2,'--g',label='Cos(x)')
plt.title('Sin(x)<->Cos(x)')
plt.legend(loc='lower left')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

```

```python
#to create multiple graph axes can be used
ax=plt.axes()
ax.plot(x,y1,'-.g',label='Sin(x)')
ax.plot(x,y2,'--r',label='Cos(x)')
ax.set(xlim=(0,12),ylim=(-1,1),xlabel='X-Axis',ylabel='Y-Axis',title='Sin(x)<->Cos(x)')
ax.legend(loc='upper right')
```

```python
#markers
plt.style.use('ggplot')
mar=['o','.','x','+','v','^','<','>','s','d']
ax=plt.axes()
ax.plot(x,y1,mar[-1],color='red',label='Sin(x)',markersize=4)
ax.plot(x,y2,mar[-2],color='green',label='Cos(x)',markersize=5)
ax.set(xlim=(0,12),ylim=(-1,1),xlabel='X-Axis',ylabel='Y-Axis',title='Sin(x)<->Cos(x)')
ax.legend(loc='upper right')
```

```python
#scatterplot
plt.figure(figsize=(5,5),dpi=150)
x=np.random.rand(100)
y=np.random.rand(100)
color=np.random.rand(100)
sizes=150*np.random.rand(100)
plt.scatter(x,y,c=color,s=sizes,alpha=0.3,cmap=plt.cm.RdGy)
plt.colorbar()
```

```python
from seaborn import load_dataset

iris=load_dataset('iris')

iris.head()
```

```python
plt.figure(figsize=(5,5),dpi=100)
colors=np.random.rand(len(iris.species))
colors[iris.species==iris.species.unique()[0]]=0.2
colors[iris.species==iris.species.unique()[1]]=0.4
colors[iris.species==iris.species.unique()[2]]=0.6

plt.scatter(iris.sepal_length,iris.sepal_width,c=colors,cmap=plt.cm.coolwarm)
plt.xlabel('Sepal_Length')
plt.ylabel('Sepal_Width')
```
